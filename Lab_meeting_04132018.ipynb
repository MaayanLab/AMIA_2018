- Systematic detection of height, weight outliers in EHR:    
    + IBM Watson Health Center
    + Methods: Fitting locally weighted regression model for age vs. height, then estimate confidence intervals (CI) at given age


- Semi-supervised encoding for outlier detection in clinical observation data
    + Shawn Murphy @ HMS/MGH
    + Train autoencoder on samples of vitals/labs/demographics data. The data points with large reconstruction error can be regarded as outliers


- City-wide opioid use among patients treated at multiple health care facilities
    + Northwestern Medicine
    + Data: HealthLNK (linked EMR from multiple medical centers in Chicago)
    + Results: 
        - ~51% of primary care patients (N=884K) had opioid records
        - 50% of opioid users visited >=2 institutions, 7% visited >=3, 0.6% visited >=4
        - pain-associated complaints are 3-fold more frequent for opioid users who visited more than 1 institutions compared to opioid users who visited only one institution

### Hyperledger Fabric (HF)
only private network
appropriate for healthcare

components:
- network 
	+ client: propose transactions
	+ peer: commit transactions
	+ ordered node: creates orders of transactions

- ledger (stores transactions and data)
	+ state DB: key-val pairs, only smart contract access data from DB
	+ transactions blocks: stores transactions
- smart contract (chaincode)
	+ read/write through its functions

- Endorsement policy: who in the network is allowed to endorse transactions
- Application interface

features:
- private channels
- fine-grained access control: e.g. access data through a certain time frame
